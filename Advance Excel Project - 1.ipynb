**App Script:**<br>
* In google sheets got Extensions > App script > Run<br>
When running script in app script . It will asks for permission (so please don't open it in incognito mode) .
1. Click on permissions.
2. Click on advanced options .
3. Click on unsafe browisng .

If you follow above conditions then only the app script will give access to run the code . So , Please follow above steps to access the app script . <br>
* **Please don't edit code in app script , because I gave access it for editable . If you edit code in app script then it might leads to ambiguous results .**

**Important:**<br>
* If you want to change Inputs and see the results . <br>
  1. After saving changes in every sheet run the app script then only we can see the perfect results. <br>
  **Example:**<br>
  1.Change inputs in sheet1(keep it active)  and run app script .Afer execution completed , the correspoing results are visible in sheet1 .<br>
  2.Change inputs in sheet2(keep it active)  and run app script .Afer execution completed , the correspoing results are visible in sheet2 .<br>
  3.Then Afer look into the sheet3 here you no need to run app script , because here I used Vlookup and Hlookup , to obtain results in sheet1 and sheet2 .
