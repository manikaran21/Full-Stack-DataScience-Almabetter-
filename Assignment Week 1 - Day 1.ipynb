## 6. What is Semi- Supervised Learning and why is it so important?


**Semi-Supervised Learning** is a machine learning technique that uses a **small portion of labeled data** and **lots of unlabeled data** to **train** the predictive **model** .<br>
Semi-Supervised learning is important because of
when we need lots of unlabeled data to train the model that data will becomes overhead while training the model and leads to the problem of more computaion cost and high time complexity . <br>
To solve the above problem semi-supervised learning comes into the picture . It can develop generalised and optimised model that can takes some portion of labeled and unlabeled data as input data to train the predictive model .

## 8. How do you treat heteroscedasticity in regression? What causes Heteroscedasticity and how to fix it?

**Heteroscedasticity:** &nbsp; Heteroscedasticity refers to situation where the variance of residuals is unequal over a range of measured values .
<br>
**Causes of heteroscedasticity in Regression Analysis:**<br>
* High error term due to presence of outliers .
* The OLS estimator remains unbaised .

**Fix the problem heteroscedasticity:**
* Transform the target variable .
* Use weighted regression which means assign a weight to each data point based on the variance of its fitted value .
<br><br>
The above two methos are used to reduce the variance of the dataset .<br>
* **A model with low variance and low bias leads to best generalised model.**
