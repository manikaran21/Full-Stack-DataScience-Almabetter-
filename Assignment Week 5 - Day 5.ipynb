## 1. What is a Model-Based Collaborative approach?

* Model-Based Collaborative  approach is a type of recommendation system that combines collaborative filtering techniques with the use of mathematical models to make predictions and recommendations.
<br>

**Steps involved in this approach:**

* **Data preparation:**
 * In this approach we need a matrix of user-item interactions or ratings , that matrix can be developed by using collected data .This matrix represents the historical interactions between users and items.

 **Matrix factorization:**
 * After developing the user-item interactions matrix . We can decompose that matrix into two lower-dimensional matrices, typically representing user and item latent factors. These latent factors capture the patterns and preferences in the data.

 **Model training:**
 * After factorization  of the matrix , the model should learn the latent factors that best explain the observed user-item interactions. This is done by minimizing a loss function that measures the difference between the actual interactions and the predicted interactions based on the learned latent factors.

 **Predictions and Recommendations:**
 * Once the model is trained and the latent factors are learned, the model can make predictions for missing or unseen interactions. It can predict how much a user might like an item based on their latent preferences and the item's latent features.
 * Recommendations are then generated by suggesting items with the highest predicted ratings.

 **Advantages of Model-Based Collaborative Filtering:**
 * It is a scalabe approach .
 * It does not suffers from cold start .
 * It can interpret the features effectively .
 * It can improve the predictions by learning .

## 2. Compare some Forecasting Techniques for Stationary and Non-stationary Time-Series.


* Stationarity refers to the property of a time series where statistical properties such as mean, variance, and autocorrelation remain constant over time.
* Non-stationary time series, on the other hand, exhibit trends, seasonality, or other patterns that evolve over time. <br>

**Forecasting techniques for Stationary time Series:**

* **ARIMA (AutoRegressive Integrated Average) Moving:**
 *  ARIMA is a widely used technique for forecasting stationary time series. It combines autoregressive (AR) and moving average (MA) components and can handle data that requires differencing to achieve stationarity. ARIMA is effective for capturing short-term patterns and autocorrelations.

* **Exponential Smoothing Methods:**

 * **simple exponential smoothing:**
   * Suitable for time series without trends or seasonality. It gives more weight to recent observations.
 * **Double exponential smoothing :**
   * Includes a trend component in addition to the level component of SES. Appropriate for time series with a linear trend but no seasonality.
 * **Triple Exponential Smoothing :**
   *  Adds a seasonal component to DES, making it suitable for time series with trends and seasonality.
* **Seasonal decomposition of time series :**
 * STL decomposes a time series into seasonal, trend, and residual components. After decomposition, forecasting can be performed separately on each component.

**Forecasting techniques for non-Stationary time series:**

* **ARIMA with Differencing:**
 * For non-stationary time series, differencing is often used to transform the data into a stationary form before applying ARIMA. The integration (I) component of ARIMA accounts for the required differencing.

* **Seasonal ARIMA (SARIMA):**
 * Extends ARIMA to handle seasonality. It includes additional seasonal AR and MA terms to account for seasonal patterns.

* **SARIMAX (Seasonal ARIMA with Exogenous Variables):**
 * Suitable when exogenous variables (variables external to the time series) influence the time series' behavior. It incorporates these variables into the forecasting process.

* **Vector Autoregression (VAR):**
 * Used for modeling multiple time series variables that are interrelated. VAR can capture dynamic interactions among variables and is suitable for non-stationary multivariate time series.

* **Vector Error Correction Model (VECM):**
 * An extension of VAR for cointegrated time series, where variables have a long-term relationship. VECM accounts for both short-term dynamics and the long-term equilibrium relationship.




## 3. Why does a Time Series have to be Stationary?



A stationary time series is a fundamental concept in time series analysis and forecasting. Stationarity refers to the property of a time series where statistical properties remain constant over time. In other words, the mean, variance, and autocorrelation structure of the data do not change as time progresses. <br>
**Few reasons for stationary times series:**
* Stationary time series exhibit consistent patterns and statistical properties are constant over time , making them easier to analyze. The relationships between variables tend to be more stable ,  patterns are more predictable and more reliable trend analysis and forecasting.
* Autocorrelation, which measures the relationship between a time series and its lagged values, is more meaningful in a stationary time series. In non-stationary series, trends and seasonality can distort autocorrelation patterns.
* Many forecasting techniques, like ARIMA, assume stationarity or work better with stationary data. Applying these techniques to non-stationary data can lead to inaccurate forecasts due to evolving patterns.
* In the presence of non-stationarity, the relationships between variables can change over time, making it difficult to build stable and consistent models. Stationarity provides a more stable foundation for modeling.
* Non-stationary time series can have trends, seasonality, and other patterns that make it difficult to interpret the underlying insights. Stationary series allow for a clearer understanding of the data generating process.



## 4. What is the difference between a Perceptron and Logistic Regression?

*  The perceptron uses a step function, which leads to binary output, while logistic regression uses the logistic (sigmoid) function, which produces a continuous probability output.

* Perceptrons are simple models that can only capture linear relationships, while logistic regression can capture both linear and nonlinear relationships through proper feature engineering.

* Perceptrons update weights based on misclassified instances, while logistic regression optimizes a loss function based on probabilities.

* Logistic regression has more consistent convergence properties, even for non-linearly separable data.

## 5. How does forward propagation and backpropagation work in deep learning?


Forward propagation and backpropagation are fundamental concepts in deep learning, specifically in training neural networks. These processes are crucial for adjusting the network's weights and biases to minimize the difference between predicted and actual outputs .

**Forward Propagation:**
* The process begins with input data given to the neural network. Each input feature is associated with a weight, and the network has bias terms for each neuron.
* In each layer, the input features are multiplied by their corresponding weights and summed up, including the bias term. This weighted sum is then passed through an activation function, which introduces nonlinearity to the model.
* The output of one layer serves as the input to the next layer, and the process continues through the hidden layers until the final output layer is reached. The final output is the prediction made by the neural network.

**Backpropagation:**

* Backpropagation is the process by which the neural network adjusts its weights and biases based on the difference between the predicted output and the actual target output.
* A loss function concludes that how far off the network's predictions are from the actual target values.
* The gradient of the loss function with respect to each parameter (weight and bias) is calculated using the chain rule of calculus. The gradients indicate the direction and magnitude of change needed in each parameter to reduce the loss.
* Starts from the output layer, the gradients are propagated backward through the layers. In each layer, the gradients are used to update the weights and biases. The gradients are adjusted by the learning rate, which determines the step size for parameter updates.
* Optimization algorithms are used to update the weights and biases in the direction that reduces the loss. These algorithms adjust the weights by subtracting a portion of the gradient times the learning rate.

Forward propagation and backpropagation are repeated for multiple iterations (epochs) over the training dataset. With each iteration, the network's parameters are gradually learned, leading to improved predictions and reduced loss.
