**a:**<br>
* Confidence tests(intervals) and Hypothesis tests are related to  inferential statistcs that means these tests are depends upon sampling distribution which is randomly selected from population.<br>
* Confidence tests(intervals) are used  to estimate a population parameter (or) that the estimated parameter falls in the range of confidence intervals or not .
* Hypothesis testing is used to prove that the statement made about a population is acceptable or not , with the significance level α .
* **One primary difference** between confidence test and hypothesis test is if the sample distribution considered for **confidence** test  then it is **centered on the observed sample statistic** and if the sample distribution considered for **hypothesis** test then it is **centered on the value in the null hypothesis**. <br>

**b:**<br>
* **Type 1 Error:**  Type 1 error occurs when we reject a null hypothesis that is actually true in the population.It is know as α .
<br>**Example :** Wrongly convicting innocent people .
* **Type 2 Error:**  Type 2 error occurs when we fail to reject a null hypothesis that is actually false in the population.It is known as β .
<br>**Example:**Letting a guilty person go free .

## 1. What is the difference between Classification and Regression when using SVM?

**SVM** is a machine learning algorithm it tries to **find a Hyperplane** in **multidimensional space** . The hyperplane must have **maximum margin**.<br>
**Margin:** It is the distance between the hyperplane and the observations closest to the hyperplane (support vectors). In SVM large margin is considered a good margin. <br>
**Support Vectors:** These are the points that are closest to the hyperplane. A separating line will be defined with the help of these data points. <br>

First , We undestand what is hyperplane and next why we need maximum margin , it will be cleared when how we are using it in regression and classification .

Let understand how we are defining hyperplane in 2 dimensional space in simple terms:
 * We are trying to find two straight lines in the 2 dimensional space(space of data points) and that the perpendicular distance between those two straight lines is maximum .
 * So, the equation of plane is **Y = W.X+b** , where W is vector normal to hyperplane and b is the offset or intercept . <br>
 Here the equation for those two lines (boundary lines) L1 and L2 in space is
    * L1 : W.X + b > 0  and L2: W.X+b < 0 <br>
    Using L1 and L2 we can derive another imaginary line , it is in between and parallel to L1 and L2 .
    * That the imaginary line is generalised line (L0) for prediction . <br>
           L0: W.X+b = 0 .
  

**SVM for Classification:**<br>
* SVM for Classification tries to find a hyperplane that best separates the two classes.<br>
                                    (or)
* In n dimensional space(assume n=2) we are trying to find two straight lines (L1 and L2) , that the lines passes through support vectors and separates the two classes with maximum distance . <br>

 * Let understand why we need maximum distance between L1 and L2  in classification: <br>
    * For predicton , we tries to find generalised line (L0) from L1 and L2 .
    * When we want to predict a new data point ,  if that the data point above the generalised line(L0) then that point belongs to positive class , if that the data point is below the generalised line(L0) then that point belons to negative class .
    * If the distance between L1 and L2 increases then the probability of the generalization of the model increases . Sine that the classifing of the data points correctly increases .
    * That the L1 and L2 must passes throungh the boundary data points of posite and negative classes respectively .
    * So , a plane with maximum margin will give more generalised model .
 * Let understand how to find maximum margin or distance between L1 and L2:<br>
    * Randomly draw all possible two lines that separates the two classes .
    *For every two lines the equation of L1 and L2 is
         L1: Y = W.X+b >=1 (+ve samples) and L2: Y = W.X+b <= 1 (-ve samples)
    * The decision rule , if the every point to be correctly classified the below condition should always be true .
                       y(i) . (W.X+b) > = 1 ;   y(i) = actual data points
    * Then equation to maximize the margin or distance is
          argmax(W,b) 2/||W|| such that y(i) (W.X+b) >= 1 , Where 2/||W|| is the distance  between support vectors .
        


**SVM for Regression:**<br>
* SVM for Classification tries to find a hyperplane that best fits the data points in a continuous space.<br>
                                    (or)
* In n dimensional space(assume n=2) we are trying to find two straight lines or boundaries (L1 and L2) , that has a maximum number of points within or closest to the decision boundary line. <br>

 * Let understand why we need maximum distance between L1 and L2  in regression: <br>
    * If the distance between L1 and L2 increases then the number of data points with in the boundaries increases then it leads to minimize the prediction error.
    * So , a plane with maximum margin will give more generalised model .

    * Let assume the distance between L1 and L2 is constant a .

## 3. a.Explain learning rate in the context of neural network models. What happens if the learning rate is too high or too low?

## b. Explain the difference between a shallow network and a deep network.

**a:**<br>
Neural networks are trained using the stochastic gradient descent optimization algorithm.In this algorithm we use learning rate (α) to update the weights of model to quickly converge into global minima .
* The learning rate is a hyperparameter that controls how much to change the model in response to the estimated error each time the model weights are updated.
* Choosing the learning rate is challenging as a value too small may result in a long training process that could get stuck in local minima , whereas a value too large may result in learning a sub-optimal set of weights too fast or an unstable training process or overshoot the convergence of the model at global optima.<br>

**b.**<br>
The main difference between deep and shallow neural networks lies in the number of layers they contain.

* Shallow neural networks, also known as single-layer neural networks, consist of only one hidden layer between the input and output layers. Shallow neural networks are  simple and have limited capacity to learn complex patterns and representations.
* Deep neural networks have multiple hidden layers stacked between the input and output layers. These networks can be quite deep, with tens or even hundreds of layers. Each layer performs computations and passes the transformed data to the next layer. Deep neural networks have a greater capacity to learn complex patterns from the data .<br>

For both the networks neurons or Perceptrons and activation fuctions are key components of the network .



## 4. What are the different types of deep neural networks?

A deep neural network (DNN) is a network with multiple hidden layers between the input and output layers .<br>
**Different Types of Neural Networks in Deep Learning:**<br>
1. **Artificial Neural Networks (ANN):**
 * Artificial Neural Networks, or ANN, is a group of multiple perceptrons/ neurons at each layer. ANN is also known as a Feed-Forward Neural network because inputs are processed only in the forward direction.<br>
 * The input layer accepts the inputs, the hidden layer processes the inputs, and the output layer produces the result. Essentially, each layer tries to learn certain weights.
 * The equation at each perceptron is:
                 f(X) = W . X + b   ,  Where (.) is dot product , W is weight matrix and X is matrix of data points.
                 O = g(f(x)) = g( W . X + b ) ,  Where g is an activation function.
2. **Convolutional Neural Networks:**
 * Convolutional Neural Networks are a special kind of neural network mainly used for image classification, clustering of images and object recognition.
 * In convolutional neural networks we use filters or filter matrix insted of weight matrix .  If we use weight matrix , it is very difficult to classify same image at different position . So , we are using the filters to detect edges in the image data .
 * So to achieve the best accuracy, deep convolutional neural networks are preferred more than any other neural network. <br>
 * The equation at each convolution is:

                 f(X) = W . X + b   ,  Where (.) is dot product , W is filter matrix and X is matrix of pixels.
                 O = g(f(x)) = g( W . X + b ) ,  Where g is an activation function.

3. **Recurrent Neural Networks**:
  * Recurrent Neural Networks are yet another variation of feed-forward networks. Here each of the neurons present in the hidden layers receives an input with a specific delay in time.
  * The Recurrent neural network mainly accesses the preceding info of existing iterations.
  * For example, to guess the succeeding word in any sentence, one must have knowledge about the words that were previously used.
