## 1. a. What is PCA? What does a PCA do?

##    b. List down the steps of a PCA algorithm.

**a.**<br>
* PCA stands for principal component analysis.
* It is a dimensionality reduction technique that transform a large set of correlated variables (high dimensional) into a lower dimensional variables, called the principal pomponents, that explains most of the variability of the original set.


**b.**<br>
**Steps of PCA algorithm:**<br>

* PCA works on a process called Eigenvalue Decomposition of a covariance matrix of a data set. The steps are as follows:

1. First, calculate the covariance matrix of a data set.
2. Then, calculate the eigenvectors of the covariance matrix.
3. The eigenvector having the highest eigenvalue represents the direction in which there is the highest variance. So this will help in identifying the first principal component.
4. The eigenvector having the next highest eigenvalue represents the direction in which data has the highest remaining variance and also orthogonal to the first direction. <br>
So, this helps in identifying the second principal component.<br>
Like this, identify the top ‘k’ eigenvectors having top ‘k’ eigenvalues to get the ‘k’ principal components.

## 3. What are the differences in Anomalies for Uniform Distribution and Normal Distribution in One-Dimensional Data?

Anomalies, also known as outliers, are data points that deviate significantly from the rest of the dataset. <br>

**Uniform Distribution:**
* Uniform distribution is characterized by constant probability density across a specified range. In the case of a one-dimensional uniform distribution, all values within that range are equally likely to occur.
* Anomalies in a uniform distribution tend to be extreme values located at the edges or beyond the defined range .<br>

**Normal Distribution:**<br>
* A normal distribution is characterized by a bell-shaped curve with the majority of data points clustering around the mean (average) value.
* Anomalies in a normal distribution are values that deviate significantly from the mean and are often several standard deviations away from it.


## 4. a. What are the Swamping and Masking problems in Anomaly Detection?

## b. What is the Change Detection problem in Anomaly Detection?

**a.**<br>
* Anomalies are the rare events, and this makes it very difficult to label these with high accuracy.
* **Swamping** is the phenomenon of labelling normal events as anomalies.
* **Masking** is the phenomenon in the clustering algorithms , when the data points belonging to different clusters gets merged into one cluster, if the number of segments (including outlier segments) in the dataset is not known. This causes the outlier cluster merged to a cluster with normal data points. Basically, the outliers are not detected.
* Swamping and Masking are more common when dataset size is large.


**b.**<br>
* Change detection is a critical aspect of anomaly detection, particularly in dynamic or evolving systems.
* The change detection problem in anomaly detection involves identifying shifts or deviations in the statistical properties of a system or dataset over time. <br>
* These shifts can signify the presence of anomalies or unusual events that require attention.

## 5. What are the steps to follow when building a text classification system?

**Steps to follow when building a text classification system:**


1. **Clarify the task:**
 * Understand the problem statement and collect the enough data to solve the problem .

2. **Data quality checks:**
 * Remove duplicates , impute missing values and outliers in the data .

3. **Exploratory Data Analysis (EDA):**
 * Analyse the dataset using EDA to uderstand the distribution of target variable and also the importance of remaining features .
4. **Text pre-processing**:
 * Text processing is necessary for the text classification systems .  Because of huge content or text leads to overhead of the machine and also it is not possible to interpret the words .So , remove the unnecessary word like stopwords , punctuations, etc .
5. **Train-test split:**
 * Split the dataset into train and test sets for training and evaluating the model.
6. **Text vectorisation:**
 * Models cannot interpret words. Instead, the words have to be converted into numbers using a process known as vectorisation. So , we can transfer that textual data into numbers using word embedding methods like one hot encoding , Bag of words , TF-IDF etc .

7. **Model selection:**
 * It’s a good idea to try out a few classification models to see which performs best with your data. You can then use performance metrics to select the most appropriate model to optimise. I did this by running a for loop which iterated over each model using the cross_validate() function.
8. **Baseline model:**
 * Select the base line model and train that model using classification algorithms .

9. **Model tuning — rectifying imbalanced data:**
 * If the data is imbalanced , then balance the data using sampling techniques and tune the parameters of the model to improve the performance of the model.
10. **Deploy trained classifier:**
 * When our model is ready then deploy that into the production and observe the performance of the model on unseen data . If our model works well then its ok otherwise improve the model or dataset and deploy into the production environment .
