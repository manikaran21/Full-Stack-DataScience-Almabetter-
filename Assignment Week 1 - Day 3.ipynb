## 1. What is multicollinearity, and how do you treat it and when is it not a problem?

* Multicollinearity occurs when two or more independent variables in a dataset have high correlation with one another .
<br>

**Fix Multicollinearity :**<br>
* Remove some of the highly correlated independent variables .
* Linearly combine the independent variables (adding) .
* Apply principal component analysis (PCA)
* Use Regularization technique .
<br>

Multicollinearity is not a problem when estimating polynomial functions or if the model focus on predictions only .


## 2. VIF(Variation Inflation Factor),Weight of Evidence & Information Value. Why and when to use?

**VIF:**
* The variance inflation factor (VIF) identifies correlation between independent variables and strength of the correlation .<br>
&nbsp;If VIF equals to 1 then there is no correlation . <br>
&nbsp; If  VIF ranges between 2-5 then indpendent variables are have very less correlation .<br>
&nbsp; If VIF ranges between 5-10 then the independent variables are moderately correlated .<br>
&nbsp; If VIF > 10 then those independent variables are highly correlated .<br>
* Variance Inflation Factor (VIF) is used to test the multicollinearity .<br>

**Weight of Evidence:**
* The weight of evidence (WoF) tells the predictive power of independent variable in relation to the dependent variable . <br>
&nbsp; WoF = ln(Percentange of events/Percentage of non-events) <br>
* The WoF value is used to know how confident we are that the feature will help us to predict the probability of an event correctly . <br>

**Information Value:** <br>
* The Information value represents the entire feature's predictive power .<br>
&nbsp; IV = SUM (WoE * ( Percentage of events - Percentage of non events)) <br>
&nbsp; From x = 1,2,3,...n  ; Where x is a feature . <br>

* Information Value helps to point out the supisious feature and feture selection .










## 3. What is tradeoff between bias and variances, and the relationship between them?

**Bias:** Bias is the difference between average predicted value and the correct target value .<br>
&nbsp; Bias = E[u1]-u ; Where u1 = Average of predicted values and u = Average of correct target values . <br>
**Variance:**The amount of variability in target function with respect to change in training data is know as Variance . <br>
&nbsp; Variance = SUM((y-u)^2) / n ; Where y = correct target value , u = Average of correct target values and n = Number of data points . <br>
**Bias-Variance tradeoff  :**<br>
* If the algorithm is too simple , it will have high bias and low variance .Similarly ,  if the algorithm is too complex , it will have low bias and high variance .<br>
* Bias and variance are complements of each other the increase of one will result in the decrease of the other and vice versa. Hence, finding the right balance of values is known as the Bias-Variance Tradeoff.
* An ideal algorithm should neither underfit nor overfit the data. The end goal of all Machine Learning Algorithms is to produce a function that has both low-bias and low-variance.


## 4. How overfitting and underfitting can affect model generalization?



**Overfitting:** The model will show good performance on the training data, poor generliazation to other data.<br>
**Underfitting:** The model will show poor performance on the training data and poor generalization to other data. <br>
* The overfitting or underfitting of the model results in poor generalised model , the poor generalised model will give inaccurate results and which is not good for the business .<br>
So , we always tends to look for best fit generalised model to get correct or desirable results .
