## 1. What are the differences between Normalization and Standardization?

Normalization and Standardization both are feature scaling techniques to normalize the data in between specific range of values .<br>
**Normalization:** <br>
  Normalization or Min-Max Scaling is used to transform features to be on a similar scale. The new point is calculated as:<br>
   **X_new = (X - X_min)/(X_max - X_min)**<br>
 This scales the range to [0, 1] . Normalization is useful when there are no outliers as it cannot cope up with them. <br>

**Standardization:**<br>
 Standardization or Z-Score Normalization is the transformation of features by subtracting from mean and dividing by standard deviation. This is often called as Z-score.<br>
   **X_new = (X - mean)/Std** <br>
 This scales the range to  [-1, 1].Afetr standardization the new data can have mean as 0 and standard deviation as 1.Standardization can be helpful in cases where the data follows a Gaussian distribution.
 Standardization does not get affected by outliers because there is no predefined range of transformed features.

## 2. Explain ANOVA and itâ€™s applications.



To make a **confident and reliable decisions**, we will need **evidence** to **support** our **approach**. This is where the concept of **ANOVA** comes into play.<br>
* **ANOVA** stands for Analysis of Variance. It compares the variation between group means to the variation within the groups. If the variation between group means is significantly larger than the variation within groups, it suggests a significant difference between the means of the groups.<br>
* ANOVA calculates an F-statistic by comparing between-group variability to within-group variability. If the F-statistic exceeds a critical value, it indicates significant differences between group means.<br>

Types of ANOVA include one-way (for comparing means of groups) and two-way (for examining effects of two independent variables on a dependent variable).
<br>
* A one-way ANOVA is used to determine how one factor impacts a response variable. For example, we might want to know if three different groups of students have different mean GPAs. To see if there is a statistically significant difference in mean GPAs, we can conduct a one-way ANOVA.<br>
* A two-way ANOVA is used to determine how two factors affect a response variable and whether or not the two factors interact with the response variable. For example, we might want to know how different types of food and how different levels of metabolism impact average weight loss. We would conduct a two-way ANOVA to find out.<br>

**Applications:**
1. An ANOVA test would most likely be used in a manufacturing plant to determine the best materials to use to build a product for a customer.
2. The ANOVA test can also be used in the business world. If a company wants to test the effectiveness of five different marketing strategies, they will use the ANOVA test.
3. Anova can be used in the healthcare industry to monitor and optimize patient care.
4. The real-time uses of ANOVA in the e-commerce industry would be to track customer satisfaction levels with different types of products.

## 1. What are the challenges of Evaluation Metrics for Imbalanced Classification Problem?
## What are the threshold Metrics,  ranking metrics and probabilistic metrics for Imbalanced Classification?

**The Challenge of Evaluation Metrics for Imbalanced Classification Problem:**<br>
* Imbalanced data(majority class >> minority class) can cause an issues in understanding the performance of a model. When evaluating performance on imbalanced data, models that only predict well for the majority class will seem to be highly performant when looking at **simple metrics such as accuracy**, On the other hand, **Balanced Accuracy** and **F1 Score** tell us that the model is performing poorly, which is what we know to be true.<br>
This means that metric choice becomes even more important in these situations.

**Threshold Metrics:**<br>
* Threshold metrics are those that quantify the classification prediction errors.They are designed to summarize the fraction, ratio, or rate of when a predicted class does not match the expected class in a given dataset.<br>
* Metrics based on a threshold and a qualitative understanding of errors . These measures are used when we want a model to minimise the number of errors.<br>
* The most widely used threshold metric is classification accuracy.<br>
  **Accuracy = Correct Predictions / Total Predictions** <br>
* And the complement of classification accuracy called classification error.<br>
  **Error = Incorrect Predictions / Total Predictions** <br>

**Ranking Metrics:**<br>
* Rank metrics are more concerned with evaluating classifiers based on how effective they are at separating classes.Metrics based on how well the model ranks the examples. These are important for many applications  where classifiers are used to select the best n instances of a set of data or when good class separation is crucial.<br>
* The most commonly used ranking metric is the ROC Curve or ROC Analysis.<br>
* A ROC curve is a diagnostic plot for summarizing the behavior of a model by calculating the false positive rate and true positive rate for a set of predictions by the model under different thresholds.<br>
 The true positive rate is the recall or sensitivity.<br>
   **TruePositiveRate = TruePositive / (TruePositive + FalseNegative)** <br>
  The false positive rate is calculated as:<br>
   **FalsePositiveRate = FalsePositive / (FalsePositive + TrueNegative)**<br>
 * The area under the ROC curve can be calculated and provides a single score to summarize the plot that can be used to compare models.
 * Again, different thresholds are used on a set of predictions by a model, and in this case, the precision and recall are calculated. The points form a curve and classifiers that perform better under a range of different thresholds will be ranked higher.<br>

**Probability Metrics:**<br>
* Probabilistic metrics are designed specifically to quantify the uncertainty in a classifierâ€™s predictions.These are useful for problems where we are less interested in incorrect vs. correct class predictions and more interested in the uncertainty the model has in predictions and penalizing those predictions that are wrong but highly confident.<br>
 * The most common metric for evaluating predicted probabilities is log loss for binary classification , or cross-entropy.<br>

## 2. How to Choose an Evaluation Metric?


**Evalution Metric:**Evaluation Metric is a value to test the performance of the model .

* The correct evaluation of learned models is one of the most important issues in pattern recognition.
* The best approach is to talk to project stakeholders and figure out what is important about a model or set of predictions. Then select a few metrics that seem to capture what is important, then test the metric with different scenarios.
* A scenario might be a mock set of predictions for a test dataset with a skewed distribution that matches our problem domain. we can test what happens to the metric if a model does well or does poorly, and so on. A few small tests can rapidly helps us to get a feeling for how the metric might perform.
* Another way to choose the metric is find the most popular metric used for the same or similar problem .<br>

Still have no idea to choose correct metric?

Here are some first-order suggestions:

* Are you want to predicting target value ? <br>
  * Is that the prediction for regression ?
       * Choose MSE or R-Square or Adjusted R-square .
  * Is that the prediction for classification ?
       * If the dataset is balanced .
         * Is the positive class more important?
           * Choose Precision-Recall
        * Are both classes important ?
           * Choose accuracy or F1 statistic .<br>

   * If the dataset is not balanced .
     * Choose accuracy or F1 score or recall or Precision using ROC AUC curve with varying thresholds .

* Are you want to group or cluster the new data point ?
  * Choose silhouette score .
