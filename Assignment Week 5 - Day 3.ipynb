## 1. a. What do you mean by null and alternative hypothesis?
## b. What is the difference between one-tailed vs. two-tailed tests?

**a.**<br>
* The null hypothesis is the claim that there’s no effect in the population.
* If the sample provides enough evidence against the claim that there’s no effect in the population (p ≤ α), then we can reject the null hypothesis. Otherwise, we fail to reject the null hypothesis.
* The alternative hypothesis (Ha) is the  claims that there’s an effect in the population.
* Some times the alternative hypothesis is the same as your research hypothesis. In other words, it’s the claim that we expect or hope will be true.
* The alternative hypothesis is the complement to the null hypothesis. Null and alternative hypotheses are exhaustive, meaning that together they cover every possible outcome. They are also mutually exclusive, meaning that only one can be true at a time.


**b.**<br>
* One and Two-Tailed Tests are ways to identify the relationship between the statistical variables.
* For checking the relationship between variables in a single direction (Left or Right direction), we use a one-tailed test. A two-tailed test is used for checking whether the relations between variables are in any direction or not.
<br>

**One-Tailed Test** <br>
* A one-tailed test is based on a uni-directional hypothesis where the area of rejection is on only one side of the sampling distribution.
* It determines whether a particular population parameter is larger or smaller than the predefined parameter. It uses one single critical value to test the data.<br>

**Two-Tailed Test**<br>
* A two-tailed test is also called a nondirectional hypothesis. For checking whether the sample is greater or less than a range of values, we use the two-tailed.
* It is used for null hypothesis testing.<br>
* The main difference between one-tailed and two-tailed tests is that **one-tailed tests **will only have **one critical region** whereas **two-tailed tests** will have **two critical regions**.

## 1. What is Latent Dirichlet Allocation (LDA) and vector space of LDA ?

 **Latent Dirichlet Allocation (LDA)**<br>
* LDA is a popular topic modeling technique to extract topics from a given corpus.
<br>
* The LDA makes two key assumptions:<br>
    * Documents are a mixture of topics .
    * Topics are a mixture of tokens (or words) .
<br>
* Firstly, LDA applies the above two important assumptions to the given corpus.
* We know the first step with the text data is to clean, preprocess and tokenize the text to words. After preprocessing the documents, we get the
preprocessed document-word matrix, in which every row is a document and every column is the tokens or the words.
* LDA converts this document-word matrix into two other matrices , one is document term matrix and other one is topic word matrix with the help of given topics .
* Finally , LDA finds the topics from the corpus, and at the same time, assigns these topics to the document present within the same corpus. <br>

**Vector space of LDA:**
* Vector space of LDA consists following parameters .
  * Dirichlet parameter 1 (α) to control per-document topic distribution .
  * Dirichlet parameter 2 (β) to control per-topic word distribution .
  * Document-topic distribution P(θ|α) .
  * Topic word distribution .
  * Word-topic assignment P(Zn|θ) , Where Z is topic assigned to the word and n is number of words in document  .
  * Observed words P(Wn|Zn,β) , Where W is word in document .



## 2. How LDA is similar to PCA?

* PCA is a dimensionality reduction technique and it is used for data having numerical values. It is the linear combination of variables from which components are derived that are used to build the model.
* PCA works by breaking or decomposing a larger value into smaller values to reduce the dimensions.
* LDA operates in the same way as PCA does. LDA is applied to the text data. It works by decomposing the corpus document word matrix (the larger matrix) into two parts (smaller matrices) the Document Topic Matrix and the Topic Word. *<br>
* Therefore, LDA like PCA is a matrix factorization technique.
<br>
* LDA breaks the corpus document word into lower-dimensional matrices. Therefore, Topic modeling and its techniques are also used for dimensionality reduction.
* So , LDA is similar to PCA interms of functionality with different applications.

## 3. What are some advantages of Content-Based Recommendation paradigm over Collaborative-Based Recommendation?


Content-Based Recommendation and Collaborative Filtering are two popular paradigms for building recommendation systems. Each has its own set of advantages and disadvantages.<br>

Here are some advantages of the Content-Based Recommendation paradigm over Collaborative-Based Recommendation:
* Content-Based Recommendation doesn't face the "cold start" problem for new items.
* It depends on item features rather than user behavior.
* It can provide personalized recommendations even for users with unique preferences.
* The recommendations made by Content-Based systems are explainable.
* It doesn't require explicit user data .
* It can recommend items that are similar to what the user has shown interest in while ensuring diversity.
* It can provide recommendations to new users based on their interest.




## 4. What is a Model-Based Collaborative approach?

* Model-Based Collaborative Filtering approach is a type of recommendation system that combines collaborative filtering techniques with the use of mathematical models to make predictions and recommendations.
<br>

**Steps involved in this approach:**

* **Data Preparation:**
 * In this approach we need a matrix of user-item interactions or ratings , that matrix can be developed by using collected data .This matrix represents the historical interactions between users and items.

 **Matrix Factorization:**
 * After developing the user-item interactions matrix . We can decompose that matrix into two lower-dimensional matrices, typically representing user and item latent factors. These latent factors capture the patterns and preferences in the data.

 **Model Training:**
 * After factorization  of the matrix , the model should learn the latent factors that best explain the observed user-item interactions. This is done by minimizing a loss function that measures the difference between the actual interactions and the predicted interactions based on the learned latent factors.

 **Predictions and Recommendations:**
 * Once the model is trained and the latent factors are learned, the model can make predictions for missing or unseen interactions. It can predict how much a user might like an item based on their latent preferences and the item's latent features.
 * Recommendations are then generated by suggesting items with the highest predicted ratings.

 **Advantages of Model-Based Collaborative Filtering:**
 * It is a scalabe approach .
 * It does not suffers from cold start .
 * It can interpret the features effectively .
 * It can improve the predictions by learning .
