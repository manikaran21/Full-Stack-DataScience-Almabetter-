## 1. What is the curse of dimensionality? Can you list some ways to deal with it?

**The Curse of Dimentionality:**<br>
* The curse of dimensionality basically means that first the performance of the model increases with the increase in the number of dimensions , but afer a certain value of number of dimensions , the performance of the model decreases with increase in number of dimensions . It refers to the fact that algorithms are harder to design in high dimensions and often have a running time exponential in the dimensions. <br>

**Ways to deal with dimensionality:**<br>
* **1.Forward-feature selection**: This method involves picking the most useful subset of features from all given features using analysis or domain knowledge or any statistical tests.

* **2.PCA:** This method help in reduction of number of features, but there is a chance of loss of information .

## 2. How do L1 and L2 regularization differ in improving the accuracy of machine learning models? Implications of using each technique and which technique is commonly preferred to boost the models accuracy rate and why?


Regularization is the technique that prevents the model from overfitting by adding extra information or penalty term to it . The penalty term is the lambda times of L1 norm (Sum of the absolute weights) or L2 norm (Sum of the squares of the weights)  . So , both L1 and L2 regularization is used to decrease the complexity of the model .<br>
* If the value of lamda equals to zero , the model is same as simple regression .<br>
* If the value of lambda increases the variance of the model decreases and bias increases then complexity of the model decreases . <br>
* If the value of lambda is too high then the model goes to underfitting the data .  So , select the appropriate lambda using cross validation . <br>

**Difference between L1 and L2 regularization:**<br>

* In L1 regularization the penalty term has the effect of forcing some of the coecient estimates to be exactly equal to zero which means there is a complete removal of some of the features for model evaluation when the tuning parameter lambda is sufficiently large. Therefore, this method also performs Feature selection .<br>
* The L2 regularization helps us to reduce only the overfitting in the model while keeping all the features present in the model. It reduces the complexity of the model by shrinking the coefficients .
<br>

**Commonly preferred technique:**<br>
* The commonly preferred technique is L1 regularization . From above reasons we conclude that **L1 regularization is more robust than L2 regularization** .

## 3. What is stratified cross-validation and when should we
## use it?

* Stratified cross validation is an extension of regular kfold cross validation but Stratified cross validation uses stratified sampling insted of random sampling . The stratified sampling ensures the training and test sets have the same proportion of the feature of interest as in the original dataset. <br>
*  Stratified cross validation specifically used for classification problems where rather than the splits being completely random, the ratio between the target classes is the same in each fold as it is in the full dataset.


## 4. Briefly explain Hyperparameter Optimization strategies and optimization strategy performance.

**Hyperparameter:** Hyperparameters are different parameter values that are used to control the learning process and have a significant effect on the performance of machine learning models.<br>
**Hyperparameter optimization** is the process of finding the right combination of hyperparameter values to achieve maximum performance on the data in a reasonable amount of time. <br>
**Methods to optimize hyperparameters:**<br>
&nbsp;**Grid Search:**<br>
* This is a widely used and traditional method that performs hyperparameter tuning to determine the optimal values for a given model.<br>
* Grid search works by trying every possible combination of parameters we want to try in our model. This means it will take a lot of time to perform the entire search which can get very computationally expensive. <br>

&nbsp;**Random Search**<br>
* This method works a bit differently, random combinations of the values of the hyperparameters are used to find the best solution for the built model.
* The drawback of Random Search is that it can sometimes miss important points (values) in the search space.
