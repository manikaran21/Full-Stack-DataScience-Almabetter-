## 2. Why can’t we use Mean Square Error (MSE) as a cost function for Logistic Regression?

**Mean Square Error (MSE)** :  The MSE is used as cost function (Regression) to update the weights to quickly convrege the model to global optima . <br>
* &nbsp;&nbsp;MSE = Sum ( ( y(i)actual - y(i)predicted)^2 ) ; i=1,2,3,...n data points .

* When we use MSE as cost function for lositic regression when the MSE loss function is plotted with respect to weights of the logistic regression model, the curve obtained is not a convex curve which makes it very difficult to find the global minimum.

* Because of MSE is suitable for continuous target values and it is not for discrete target values . But in Logistic regression we need to predict binary discrete values .

* If we don't get global optima , it is very diffict to train the model and the obtained model is also inefficient .

* Therefore  we can't use Mean Square Error (MSE) as a cost function for Logistic Regression .



## 3. What do you understand by Pruning in a Decision Tree and Why do we require Pruning in Decision Trees? Explain.

**Decision tree:** <br>
* Decision trees are a Supervised ML algorithm with a tree based prediction method. They are fairly unique in the world of Machine Learning since in that there is no reason to scale features, or do any type of standardization required.
* However, they are prone to over fitting due to the nature of making a split at each node, and let assume N as the number of data points, k as the Number of features and d as Depth of the tree .
Time complexity to learn a single node of a decision tree is done by calculating the sorting and updating the impurities as we go.If we go through this it leads to more time complexity and overfitting the model .<br>
* Hence needs some form of **regularizatio**n to **reduce the complexity**.<br>

**Pruning is the method to reduce over fitting:** <br>
&nbsp;&nbsp;&nbsp;**Pre-Pruning:**
When working with tree based methods, there is a much more direct way to control the complexity, which is with the number of decision nodes.
A decision tree which subdivides all the way to individual data points is maximally over fit.This can be controlled either directly (say by limiting the number of leaves) or by imposing a maximum depth.
This idea (stopping early when some criteria isn’t met) is called pre-pruning.
<br>
&nbsp;&nbsp;&nbsp;**Post-Pruning:**
In this case, we start with a complex tree, and trim backwards
Simplest way is reducing error pruning: iterate over leaf nodes removing those that can be removed without hurting the predictive accuracy on a withheld validation set.



## 4. What do you understand about Information Gain? Also, explain the mathematical formulation associated with it.


**Information Gain:**<br>

* Information gain is the difference between the entropy of a data segment before the split and after the split i.e, reduction in impurity due to the selection of an attribute. <br>
* The high difference represents high information gain.
* Higher the difference implies the lower entropy of all data segments resulting from the split.
* Thus, the higher the difference, the higher the information gain, and the better the feature used for the split.
<br>

**Entropy:**<br>

* Entropy is the degree of uncertainty, impurity or disorder of a random variable, or a measure of purity. It characterizes the impurity of an arbitrary class of examples.

&nbsp;&nbsp;&nbsp; E = - SUM(p(i) log(p(i)))  ; where i = 1 ,2,3,...n classes . <br>

**MatheMatical Formula:**<br>
* **Information Gain = E(S1) – E(S2)** <br>


**E(S1)**denotes the entropy of data belonging to the node before the split.<br>

**E(S2)** denotes the weighted summation of the entropy of children nodes by considering the weights as the proportion of data instances falling in specific children nodes.<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;E(S2) = SUM( (C(i)/N) * E) ; where E entropy of child node after split and C(i) is the numper of points in class of split.
