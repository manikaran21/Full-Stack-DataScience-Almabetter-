## 1. What is “K” in the KNN Algorithm, how to choose the optimal value of K in the KNN Algorithm, and why is the odd value of “K” preferred over even values in the KNN Algorithm?

* **1.** KNN(K-nearest neighbours) is a supervised learning and non-parametric algorithm that can be used to solve both classification and regression problem statements.
* **2.** K represents the number of nearest neighbours we want to select to predict the class of a given item, which is coming as an unseen dataset for the model.
* **3.** There is no straight forward method to find the optimal value of K in the KNN algorithm.
<br>The optimum value of K for KNN is highly dependent on the data itself.
<br>

**Suggestions:**<br>
* **Square Root Method:** Take the square root of the number of samples in the training dataset and assign it to the K value.<br>
* **Cross-Validation Method:** We should also take the help of cross-validation to find out the optimal value of K in KNN. Start with the minimum value of k i.e, K=1, and run cross-validation, measure the accuracy, and keep repeating till the results become consistent.
<br>

* **4.** The odd value of K should be preferred over even values in order to ensure that there are no ties in the voting. If the square root of a number of data points is even, then add or subtract 1 to it to make it odd.







## 2. How can you relate KNN Algorithm to the Bias-Variance tradeoff?

**Problem with having too small K:**<br>
* The major concern associated with small values of K lies behind the fact that the smaller value causes noise to have a higher influence on the result which will also lead to a large variance in the predictions. <br>

**Problem with having too large K:**<br>
* The larger the value of K, the higher is the accuracy. If K is too large, then our model is under-fitted. As a result, the error will go up again.
So, choosing k to a large value may lead to a model with a large bias(error).<br>

**The effects of k values on the bias and variance:**
* As the value of k increases, the bias will be increases and the boundary becomes smoother .
* As the value of k decreases, the variance will increases .
* So, there is a tradeoff between overfitting and underfitting and we have to maintain a balance while choosing the value of K in KNN.  <br>

* Therefore, **K** should **not** be **too small** or **too large**.

## 3. Why Naive Bayes is called Naive and how does the Naive Bayes classifier work?

**Naive Bayes Classifier:**<br>
 * Naive Bayes classifier is a supervised learning algorithm, which is based on Bayes theorem and used for solving classification problems.<br>
* **Naive:** It is called Naive because it assumes that the occurrence of a certain feature is independent of the occurrence of other features. <br>

**Working of Naive Bayes Classifier:**
   * It uses bayes theorem , P(A|B) = P(B|A)P(A)/P(B) <br>
   Where ,<br>
   P(A|B) is Posterior probability:  Probability of hypothesis A on the observed event B.<br>
   P(B|A) is Likelihood probability: Probability of the evidence given that the probability of a hypothesis is true.<br>
   P(A) is Prior Probability: Probability of hypothesis before observing the evidence.<br>
   P(B) is Marginal Probability: Probability of Evidence. <br>

2. Generate Likelihood table by finding the probabilities of given features.
3. Now, use Bayes theorem to calculate the posterior probability.
